# Analysis of Covid 19 Data from Kaggle

```python
# Importing the Packages

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

```python
# Importing the Dataset

data = pd.read_csv(r"E:\Python for Data Science and Machine Learning\Extra data for Machine Learning\covid_19_dataset.csv", 
                   encoding = "unicode_escape")
```

```python
ax = sns.relplot(data=data,
x="total_cases",y="total_recovered", hue = "total_cases_per1m",
estimator=sum,ci=False)
ax.figure.set_size_inches(10, 8)
```

```python
ax = sns.relplot(data=data,
x="total_cases",y="total_deaths", hue = "serious_critical",
estimator=sum,ci=False)
ax.figure.set_size_inches(10, 8)
```

```python
ax = sns.relplot(data=data,
x="new_cases",y="active_cases", hue = "total_cases",
estimator=sum,ci=False)
ax.figure.set_size_inches(18, 15)
```

```python
ax = sns.relplot(data=data,
x="new_deaths",y="total_deaths", hue = "total_cases",
estimator=sum,ci=False)
ax.figure.set_size_inches(10, 8)
```

```python
sns.pairplot(data)
```

```python
ax = sns.relplot(x = "serious_critical", y = "total_recovered", kind = "line", data=data)
ax.figure.set_size_inches(18, 8)
```
